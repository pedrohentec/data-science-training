# Removendo imóveis de tipos comerciais

```python
# Primeiro precisamos saber quais são os nomes da coluna Tipo
''' Utilizando o método "unique", ele retorna um array com todos os nomes'''

dados.Tipo.unique()
```

```python
# Após isso, criamos uma lista com todos os imóveis COMERCIAIS

imoveis_comerciais = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial', 'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio', 'Pousada/Chalé',
                      'Hotel', 'Indústria']
```

```python
# Agora vamos remover alguns dos tipos que tem dentro da lista de imoveis comerciais
'''O método "query", é usado para selecionar linhas de dataframe com base em uma condição especifica'''
# No exemplo abaixo estamos selecionando apenas os imóveis_comerciais na coluna Tipo

dados.query("@imoveis_comerciais in Tipo")
```

```python
''' Seguindo o mesmo procedimento acima, podemos fazer o inverso, NÃO selecionando
os imoveis_comerciais'''
# Fazendo uma rápida observação só temos de diferente o uso do "not"

imoveis_residenciais = dados.query("@imoveis_comerciais not in Tipo")
imoveis_residenciais
```

```python
# E agora que temos somente os imóveis residenciais podemos fazer o uso do gráfico e ver a média

df_preco_medio = imoveis_residenciais.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_preco_medio.plot(kind="barh", figsize=(10,6), color = "black");
```

# Removendo registros incosistentes

Iremos remover todo e qualquer tipo de valor que seja igual a 0 ou dados que não relevantes para o nosso dataframe

```python
imoveis_residenciais
```

```python
# Remova apartamentos que possuem valor de aluguel igual a 0

# Primeiro vamos SELECIONAR as linhas em que as colunas valor ou condimínio serão iguais a 0
# o sinal "|" é lido como "ou"
imoveis_residenciais.query("Valor == 0 | Condominio == 0")
```

```python
# Capturando os indices desses valores 0 com o ".index" e salvando numa variavel

registros_a_remover = imoveis_residenciais.query("Valor == 0 | Condominio == 0").index
```

```python
# E agora vamos usar o método ".drop()" que é usado para remover linhas e colunas

'''Passando além do nome da variável para removermos, usamos tambem o "axis", onde se passar o valor 0 remove linhas e 1 remove colunas
E passando o parâmetro "inplace", estamos atribuindo essa nova alteração na variável sem precisar declarar novamente, por exemplo:
imoveis_residenciais = imoveis_residenciais...'''

imoveis_residenciais.drop(registros_a_remover, axis=0, inplace=True)
```

```python
# E agora vamos conferir se algum dado em zero ficou em nosso DataFrame

imoveis_residenciais.query("Valor == 0 | Condominio == 0")
```

```python
# Estando tudo limpo fazemos a visualização do mesmo

imoveis_residenciais
```

# Manipulação de Dados

Criação de colunas numéricas

```python
# valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;

url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv"
df = pd.read_csv(url, sep=";")
df.head()
```

```python
df["Valor_por_mes"] = df["Valor"] + df["Condominio"]
df.head()
```

```python
# valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

df["Valor_por_ano"] = df["Valor_por_mes"] * 12 + df["IPTU"]
df.head()
```

Criação de colunas categóricas

```python
df.head()
```

```python
'''Descrição: essa coluna deve possuir uma sumarização das principais informações
 dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;'''

df["Descricao"] = df["Tipo"] + " em " + df["Bairro"] + " com " + df["Quartos"].astype(str) +\
" quarto(s) e " + df["Vagas"].astype(str) + " vagas de garagem."

df.head()
```

```python
'''Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes,
sem se importar com a quantidade.'''

df.head()
```

```python
df["Possui_suite"] = df["Suites"].apply(lambda x: "Sim" if x > 0 else "Não")
df.head()
```

```python
df.to_csv("arquivo_final.csv", index=True, sep=";")
```
