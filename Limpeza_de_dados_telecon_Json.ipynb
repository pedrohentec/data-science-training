```python
import pandas as pd
data = pd.read_json('dataset-telecon.json')
```

```python
# Achatando a coluna 'cliente' em colunas individuais
cliente_df = data['cliente'].apply(pd.Series)

# Concatenando as colunas achatadas com o dataframe original
data = pd.concat([data, cliente_df], axis=1)

# Removendo a coluna original 'cliente'
data.drop('cliente', axis=1, inplace=True)

# Exibindo as primeiras linhas do dataframe atualizado
data.head()
```

```python
# Achatando a coluna 'telefone' em colunas individuais
telefone_df = data['telefone'].apply(pd.Series)

# Concatenando as colunas achatadas com o dataframe original
data = pd.concat([data, telefone_df], axis=1)

# Removendo a coluna original 'telefone'
data.drop('telefone', axis=1, inplace=True)

# Achatando a coluna 'internet' em colunas individuais
internet_df = data['internet'].apply(pd.Series)

# Concatenando as colunas achatadas com o dataframe original
data = pd.concat([data, internet_df], axis=1)

# Removendo a coluna original 'internet'
data.drop('internet', axis=1, inplace=True)

# Achatando a coluna 'conta' em colunas individuais
conta_df = data['conta'].apply(pd.Series)

# Concatenando as colunas achatadas com o dataframe original
data = pd.concat([data, conta_df], axis=1)

# Removendo a coluna original 'conta'
data.drop('conta', axis=1, inplace=True)

# Exibindo as primeiras linhas do dataframe atualizado
data.head()
```

```python
# Achatando a coluna 'cobranca' em colunas individuais
cobranca_df = data['cobranca'].apply(pd.Series)

# Concatenando as colunas achatadas com o dataframe original
data = pd.concat([data, cobranca_df], axis=1)

# Removendo a coluna original 'cobranca'
data.drop('cobranca', axis=1, inplace=True)

# Exibindo as primeiras linhas do dataframe atualizado
data.head()
```

```python
# Verificando a quantidade de valores faltantes por coluna
missing_values = data.isnull().sum()

missing_values
```

```python
# Filtrando e exibindo as linhas onde 'tempo_servico' possui valores faltantes
data[data['tempo_servico'].isnull()]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotando a distribuição de 'tempo_servico'
plt.figure(figsize=(10, 6))
sns.histplot(data['tempo_servico'], kde=True, bins=30)
plt.title('Distribuição de tempo_servico')
plt.xlabel('Tempo de Serviço')
plt.ylabel('Contagem')
plt.grid(True)
plt.show()
```

```python
# Substituindo os valores faltantes em 'tempo_servico' pela mediana
data['tempo_servico'].fillna(data['tempo_servico'].median(), inplace=True)

# Verificando novamente a quantidade de valores faltantes
missing_values_updated = data.isnull().sum()

missing_values_updated
```

```python
# Filtrando e exibindo as linhas onde 'contrato' possui valores faltantes
data[data['contrato'].isnull()]
```

```python
# Plotando a distribuição de valores da coluna 'contrato'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='contrato', order=data['contrato'].value_counts().index)
plt.title('Distribuição de Valores da Coluna Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Contagem')
plt.grid(axis='y')
plt.show()
```

```python
# Substituindo os valores faltantes em 'contrato' pela moda
data['contrato'].fillna(data['contrato'].mode()[0], inplace=True)

# Verificando novamente a quantidade de valores faltantes
missing_values_updated_2 = data.isnull().sum()

missing_values_updated_2
```

```python
# Substituindo os valores faltantes em 'faturamente_eletronico' pela moda
data['faturamente_eletronico'].fillna(data['faturamente_eletronico'].mode()[0], inplace=True)

# Verificando novamente a quantidade de valores faltantes
missing_values_updated_3 = data.isnull().sum()

missing_values_updated_3
```

```python
# Plotando a distribuição de 'mensal' e 'Total'
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Distribuição de 'mensal'
sns.histplot(data['mensal'], kde=True, ax=axes[0])
axes[0].set_title('Distribuição de Mensal')
axes[0].set_xlabel('Mensalidade ($)')
axes[0].set_ylabel('Contagem')
axes[0].grid(True)

# Distribuição de 'Total'
sns.histplot(data['Total'], kde=True, ax=axes[1])
axes[1].set_title('Distribuição de Total')
axes[1].set_xlabel('Total ($)')
axes[1].set_ylabel('Contagem')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Obtendo estatísticas descritivas para as colunas 'mensal' e 'Total'
mensal_description = data['mensal'].describe()
total_description = data['Total'].describe()

mensal_description, total_description
```

```python
# Convertendo a coluna 'Total' para numérico e tratando possíveis erros de conversão
data['Total'] = pd.to_numeric(data['Total'], errors='coerce')

# Verificando novamente a quantidade de valores faltantes após a conversão
missing_values_total_conversion = data.isnull().sum()

missing_values_total_conversion
```

```python
# Substituindo os valores faltantes em 'mensal' pela mediana
data['mensal'].fillna(data['mensal'].median(), inplace=True)

# Substituindo os valores faltantes em 'Total' pela mediana
data['Total'].fillna(data['Total'].median(), inplace=True)

# Verificando novamente a quantidade de valores faltantes
missing_values_updated_5 = data.isnull().sum()

missing_values_updated_5
```

```python
# Verificando se há linhas duplicadas no dataset
duplicated_rows = data[data.duplicated()]

# Quantidade de linhas duplicadas
num_duplicated_rows = duplicated_rows.shape[0]
num_duplicated_rows
```

```python
# Removendo as linhas duplicadas
data.drop_duplicates(inplace=True)

# Verificando novamente a quantidade de linhas duplicadas após a remoção
num_duplicated_rows_after_removal = data[data.duplicated()].shape[0]
num_duplicated_rows_after_removal
```

```python
# Plotando boxplots para as colunas numéricas
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Boxplot para 'tempo_servico'
sns.boxplot(y=data['tempo_servico'], ax=axes[0])
axes[0].set_title('Boxplot de Tempo de Serviço')
axes[0].set_ylabel('Meses')

# Boxplot para 'mensal'
sns.boxplot(y=data['mensal'], ax=axes[1])
axes[1].set_title('Boxplot de Mensalidade')
axes[1].set_ylabel('Valor ($)')

# Boxplot para 'Total'
sns.boxplot(y=data['Total'], ax=axes[2])
axes[2].set_title('Boxplot de Total')
axes[2].set_ylabel('Valor ($)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Aplicando a transformação logarítmica à coluna 'Total'
data['Total_log'] = np.log(data['Total'] + 1)

# Plotando a distribuição original e a distribuição após a transformação logarítmica
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Distribuição original de 'Total'
sns.histplot(data['Total'], kde=True, ax=axes[0])
axes[0].set_title('Distribuição Original de Total')
axes[0].set_xlabel('Total ($)')
axes[0].set_ylabel('Contagem')
axes[0].grid(True)

# Distribuição após transformação logarítmica
sns.histplot(data['Total_log'], kde=True, ax=axes[1])
axes[1].set_title('Distribuição de Total após Transformação Logarítmica')
axes[1].set_xlabel('log(Total)')
axes[1].set_ylabel('Contagem')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Colunas a serem normalizadas
columns_to_scale = ['tempo_servico', 'mensal', 'Total_log']

# Aplicando a normalização
scaler = StandardScaler()
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# Exibindo as primeiras linhas do dataframe após a normalização
data.head()
```
