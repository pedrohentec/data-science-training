# Desagrupar Dados

Ao importar nossos dados, podemos observar que da coluna "descricao_local" até a coluna "preco", temos informação que estão entre colchetes, ou seja, agrupadas em listas.

Vamos estar desagrupando para podermos trabalhar e analisá-los melhor.

Através do método `explode()`.

O `explode()` transforma cada elemento de uma lista em uma linha. Nele passando o parâmetro das colunas que desejamos fazer essa transformação

```python
colunas = list(dados.columns)
colunas
```

```python
# Observado que precisamos só trabalhar da coluna descricao_local, vamos utilizar o indice:

dados = dados.explode(colunas[3:])
dados
```

# Convertendo Dados Numéricos

Foi observado que os tipos de todos os dados estão em `object`, entretanto, para trabalharmos com probilidade e estatística, precisamos ter valores numéricos, com isso vamos utilizar o método `astype()` para fazermos a conversão do tipo `object` para `int` ou `float`

Para isso precisamos importar a biblioteca Numpy, para podermos converter em tanto `int64` quanto `float64`, que são tipos especificos de valores com precisão de 64 bits.

```python
import numpy as np
```

```python
# Nesse caso estamos fazendo a conversão de uma única coluna

dados["max_hospedes"] = dados["max_hospedes"].astype(np.int64)
```

```python
colunas_numericas = ["quantidade_banheiros",	"quantidade_quartos",	"quantidade_camas"]

dados[colunas_numericas] = dados[colunas_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
# Convertendo em valor float

dados["avaliacao_geral"] = dados["avaliacao_geral"].astype(np.float64)
```

```python
dados.info()
```

Porém agora, temos as últimas 3 colunas:

- taxa_deposito
- taxa_limpeza
- preco

Essas colunas, possuem sifrão e virgulas, e não podemos usar o `astype()`, por isso, precisamos tratar isso antes de fazer a conversão desses valores e passar para `float`

# Números em Strings

Agora vamos trabalhar com a coluna de preço, onde vamos substituir o tipo de dos dados para float, porém fazendo algumas manipulações para que isso seja possível:

```python
# Utilizando o replace, strip e lambda

dados["preco"] = dados["preco"].apply(lambda x: x.replace("$", "").replace(",", "").strip())
```

```python
# Aplicando o astype para tentar a conversão do tipo, de string para float

dados["preco"] = dados["preco"].astype(np.float64)
```

```python
dados.info()
```

# Tratando colunas múltiplas

Anteriormente usamos o `apply` que faz a aplicação de uma única colunas.

Agora vamos usar o `applymap`, quando tratamos da conversão de um dataframe, no qual passaremos por cada elemento.

Vamos manipular as colunas:

- taxa_deposito
- taxa_limpeza

```python
dados[["taxa_deposito", "taxa_limpeza"]] = dados[["taxa_deposito", "taxa_limpeza"]].applymap(lambda x: x.replace("$", "").replace(",", ""))
```

```python
dados[["taxa_deposito", "taxa_limpeza"]] = dados[["taxa_deposito", "taxa_limpeza"]].astype(np.float64)
```

```python
dados.info()
```

# Análise Textual

Para que possa ser feito uma boa análise textual, podemos realizar a manipulação destes textos de cada colunas para criar um tipo de tokenixação simples.

`Tokenização` é a divisão do texto em `tokens`, sendo unidades menores, podendo ser palavras, sílabas, conjuntos de palavras e etc.

Por se tratar de tokenização simples, iremos fazer a separação por palavras. (tokens sera cada palavra).

Para isso vamos estar iniciando a nossa visualização da coluna `"descricao_local"` e darmos continuidade:

```python
dados["descricao_local"]
```

```python
# Vamos converter as letras maiúsculas em minúsculas, com isso vamos converter todos esses textos em string e usar o lower()
dados["descricao_local"] = dados["descricao_local"].str.lower()
```

```python
dados["descricao_local"]
```

Com isso podemos retirar os caracteres que não traz nenhum tipo de informação.

# Remoção de caracteres no texto

Agora vamos estar removendo alguns caracteres que não são úteis e que não nenhum valor para a nossa análise de dados textual.

Para isso, vamos utilizar o replace com regex.
Mas antes precisaremos transformar isso em um tipo de string, ficando dessa forma:

```python
dados["descricao_local"] = dados["descricao_local"].str.replace("[^a-zA-Z0-9\-\']", " ", regex=True)
```

Mas que parâmetro voce passou nesse regex ?

- `^` Usado quando queremos negar os próximos valores, tudo que colocar em seguida, ele não vai selecionar para aplicar o replace

- `a-z` todas as letras, do "a" ao "z", minúsculas
- `A-Z` todas as letras, do "A" ao "Z", maiúsculas
- `0-9` todo os números, de 0 a 9
- `\-` hífen
- `\'` apóstrofo

Ou seja, tudo que passamos depois do `^` ele NÃO vai selecionar para aplicar em nosso Regex

Com isso, temos nossa coluna "descricao_local", apenas com esses dados que foram selecionados para NÃO remover.

Mas e se precisássemos remover os hífen `-` que não compões palavras ? Como poderiamos fazer isso ?

Utilizando um código semelhante, porém passando um novo parâmetro para fazermos a seleção:

```python
dados["descricao_local"] = dados["descricao_local"].str.replace("(?<!\w)-(?!\w)", " ", regex=True)
```

- `(?<!\w)`: uma "lookbehind" negativa que verifica se há um caractere de palavra antes do hífen ou se há apenas um espaço em branco, resultando em "verdadeiro" se não houver nada

- `(?!\w)`: uma lookahead negativa que verifica se há um caractere de palavra depois do hífen.

Com isso, podemos garantir que se não houver nenhuma palavra antes e nem depois vamos estar substituindo por espaço em branco

```python
dados["descricao_local"]
```

# Tokenização de Strings

Para realizarmos a tokenização de strings, podemos simplesmente utilizar a função `split()`. Ela faz com que seja separado cada palavra após um espaço, deixando nossa coluna `descricao_local` da seguinte forma:


```python
dados["descricao_local"] = dados["descricao_local"].str.split()
dados["descricao_local"]
```

```python
# Agora vamos fazer a mesma coisa com a coluna "comodidades"
dados["comodidades"]
```

```python
# Usando seguinte código para remover a abertura e o fechamento das chaves e as aspas duplas

dados["comodidades"] = dados["comodidades"].str.replace('|{|}|\"', "", regex=True)
dados["comodidades"]
```

```python
# Passando como parâmentro na função split para especificar que o separador é a virgula
dados["comodidades"] = dados["comodidades"].str.split(",")
```

```python
dados.head()
```

# Manipulação de dados Temporais

Vamos fazer o agrupamento da coluna `data`, para vermos quantas vagas disponivel há em cada mês. Por exemplo, sabemos que nosso dataframe possui 4 colunas, e trabalharemos com a coluna `data` e `vaga_disponivel`, com elas vamos fazer um agrupamento para saber quanto imoveis estão disponível por cada mês:

Ex.:
- Janeiro = 100
- Fevereiro = 1234
- ...

```python
dados_imoveis["data"]
```

```python
# Vamos utilizar o método strftime para converter os valores de datas para o formato que quisermos
# adicionamos o "dt" para chamar o tipo datetime

dados_imoveis["data"].dt.strftime("%Y-%m")
```

```python
# Com isso agora, vamos fazer o agrupamento por cada mês
# Vamos adicionar ao final a função "sum()" para fazermos a soma da coluna vaga_disponivel, onde False=0 e True=1
meses_agrupados = dados_imoveis.groupby(dados_imoveis["data"].dt.strftime("%Y-%m"))["vaga_disponivel"].sum()
```

```python
meses_agrupados
```
