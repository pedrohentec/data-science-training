Fizemos o processo para descobrir a média dos filmeID 1 e 2, porém e se precisássemos ver de TODOS os filmes, iriamos fazer um por um?

Claro que não!

De início, vamos fazer um agrupamento do dataframe filmes da coluna filmeID;

Depois de agrupar, vamos descobrir a média de nota de cada filmes:

```python
# Agrupando a coluna filmeID do dataframe Filmes

notas.groupby("filmeID")
```

```python
# Visualizando a média da nota de cada filme, extraindo somente a coluna "nota"

medias_por_filmes = notas.groupby("filmeID").mean()["nota"]
medias_por_filmes.head()
```

```python
# Visualizando a média de notas por filme em um gráfico de histograma

medias_por_filmes.plot(kind="hist");
```

```python
# Visualizando a média de notas por filme em um gráfico de histograma utilizando seaborn

sns.histplot(x=medias_por_filmes);
```

Como se pode ver, há uma diferença na visualização, pois ambos os gráficos acima são em histograma, utilizando a plotagem do Pandas, ele irá retornar uma quantidade diferente de linhas na vertical, porém no seaborn é possível setar a quantidade para que se tenha a mesma visualização, passando o parâmetro `bins=quantidade de linhas`.

Porém lembre-se de não passar uma visão errada sobre a análise do gráfico, por padrão o próprio seaborn faz essa seleção de quantidade para ser apresentada.

```python
sns.histplot(x=medias_por_filmes, bins=10);
```

```python
# Usando a plotagem no Seaborn
# Média de notas por filmes em boxplot

sns.boxplot(x=medias_por_filmes)
```

# **Visualizando Gráficos por categoria**


```python
# Contagem de quantidade de idiomas

filmes_tmdb["original_language"].value_counts()
```

Acima temos a visualização de uma Series, por mais semelhante que seja a visualização de uma tabela, nessa situação temos os indices `(index)` ou seja, quando possamos que o valor de `en` é igual a `4505`, temos como se o `en` fosse o `index` e o `4505` fosse o `valor`. Por exemplo:


```python
filmes_tmdb["original_language"].value_counts().index
```

```python
filmes_tmdb["original_language"].value_counts().values
```

Com essa visualização acima podemos ter como afirmação que a visualização inicial é uma série. Porém caso queiramos ter como visualização em DataFrame, podemos utilizar a função `to_frame()`:


```python
# Tendo como a conversão da visualização em DataFrame

filmes_tmdb["original_language"].value_counts().to_frame()
```

Agora vamos fazer o reset do index, pois nao queremos que a lingua do filme seja o nosso index, com isso teremos quer fazer o reset e vamos salvar em uma nova variável:


```python
# Salvando os dados em uma nova variável

contagem_de_lingua = filmes_tmdb["original_language"].value_counts().to_frame().reset_index()
contagem_de_lingua.head()
```

```python
# Renomeando as colunas

contagem_de_lingua.columns = ["original_language", "total"]
contagem_de_lingua.head()
```

# **Plotagem de Categorias**

No site oficial do Seaborn é possível identificar alguns exemplos onde é possível fazer a comparação de dados categóricos, ou seja, comparar as categorias entre elas. Por exemplo:


```python
# Plotando método de comparação no Seaborn utilizando gráfico de barras na vertical

sns.barplot(data = contagem_de_lingua, x="original_language", y="total");
```

```python
# Melhorando a visualização dos dado alterando o tamanho do gráfico

ax = sns.barplot(data = contagem_de_lingua, x="original_language", y="total")
ax.figure.set_size_inches(10, 8)
```

Beleza, mais vimos que precisamos separar cada coluna para poder fazer essa plotagem, será que não existe uma forma mais rápida e "simples" de ser executada essa plotagem de categoria ?

E sim, é possível! Com o Seaborn, voce pode passar no eixo X a categoria, puxar o data = ao dataframe inicial (que no nosso caso é o "filmes_tmdb") e usamos um parâmetro para fazer a contagem, mas para passar esse parâmetro precisamos usar o `kind="count"`, ou seja, o nosso código ficará semelhante a isto:

`sns.catplot(data=filmes_tmdb, x="original_language", kind="count")`

E o que estamos informando aqui? Como fazemos a leitura dessa linha ?

" *-Seaborn, plota pra mim a categoria da coluna "original_language" do DataFrame(data) "filmes_tmdb", e faça em formato de contagem "kind='count'", ou seja pega todos os valores únicos de idiomas e plote em quantidade que os mesmos aparecem na coluna ("original_language").*"

Antes de fazer esse processo da linha de código acima, precisamos observar qual a versão do nosso seaborn, pois para utilizar o `catplot`, precisamos estar acima da versão 0.9.0, desta forma vamos verificar qual a versão que estamos utilizando dentro do espaço disponibilizado pelo Google Colab.

# **Passando uma mensagem através de visualização**

Quando estamos trabalhando na área de dados, temos a missão de transmitir uma "mensagem" através das análises, seja com gráficos ou até mesmo com artigos explicativos.

Nessa situação onde estamos aprendendo sobre plotagens, temos como missão entender e enviar uma mensagem através daquilo que estamos analisando, ou seja, decidir que gráficos seria o mais recomendado, legendas, tamanho, tipo de dados, e etc.

Com isso, vamos entender um pouco sobre qual a melhor recomendação para a nossa situação que se enquadra em uma situação real.

```python
# Testando plotagem de "Pizza" (torta em Inglês)
# Obs.: Esse tipo de gráfico não possui no Seaborn, utilizaremos o Matplotlib

plt.pie(contagem_de_lingua["total"], labels=contagem_de_lingua["original_language"]);
```

Como podemos observar o gráfico de formato de Pizza (pie (torta em inglês)), não é muito recomendado, pois a olho nu, fica "quase impossível" de fazer uma "leitura" dessa plotagem.

Tivemos o gráfico acima e o gráfico de barras, porém tanto um como o outro não servem para transmitir uma mensagem nitída, que é a comparação de quantidade de idiomas dos filmes, nessa situação estamos apenas exercitando algumas práticas de plotagem, porém quando temos como definido o que realmente desejamos transmitir, fica mais "facíl", porque podemos pesquisar ou encontrar modelos de plotagem que se adequariam melhor para a situação.

Porém como dito acima, como podemos entender melhor a mensagem que queremos transmitir ? Nessa situação fica evidente que a língua inglesa é disparada a que mais esta nos filmes, com isso vamos fazer a comparada dela com as outras linguas.

```python
# Pegando o total de valores das lingua na coluna "original_language"
total_de_lingas = filmes_tmdb["original_language"].value_counts()

# Somando o total geral das líguas
total_geral = total_de_lingas.sum()

# Coletando o total de valores da lingua inglesa
total_ingles = total_de_lingas.loc["en"]

# Coletando o resto dos idiomas
total_do_resto = total_geral - total_ingles

# Printando a quantidade de valores de ingles e do resto dos idiomas
print(f"Valor Total Inglês = {total_ingles}",
      f"\nValor Outros Idiomas = {total_do_resto}")
```

```python
# Criando um dataframe com base nas informações acima

# Colunas = "lingua" e "total"
# Coluna "lingua" = "ingles" e "outros"
# Coluna "total" = os valores acima de (seguindo a ordem das colunas acima) "total_ingles" e "total_do_resto"


dados_linguas = {
    "lingua" : ["inglês", "outros"],
    "total" : [total_ingles, total_do_resto]
}

dados_linguas
```

```python
# Criando o DataFrame com base na construção acima

dados_linguas = pd.DataFrame(dados_linguas)
dados_linguas
```

```python
# Plotagem em barras usando o seaborn

sns.barplot(data=dados_linguas, x="lingua", y="total");
```

Com isso podemos finalizar a mensagem que queriamos transmitir, qual seja a comparação de quantidade de filmes de lingua inglesa com os demais idiomas.

# **Visualizando a categoria "Outros"**

Vimos a questão da comparação entre o idioma inglês com os outros idiomas, porém se quiséssemos entender como isto esta sendo distribuido, como poderiamos fazer isto ?

- Primeiro vamos selecionar os outros idiomas que sejam diferentes de inglês

```python
# Selecionando idiomas diferentes de inglês

filmes_sem_lingua_em_ingles = filmes_tmdb.query("original_language != 'en'")
filmes_sem_lingua_em_ingles.head()
```

```python
# Fazendo a contagem de valores de outros idiomas

total_de_lingua_de_outros_idiomas = filmes_sem_lingua_em_ingles["original_language"].value_counts()
total_de_lingua_de_outros_idiomas
```

Depois de separmos quais os dados que vamos trabalhar para plotar ou seja, os filmes SEM o idioma inglês, vamos agora fazer a plotagem do mesmo.

```python
sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count");
```

# **Refinando visualizações**

Com base no gráfico acima, vamos melhorar a visualização dele para termos uma análise visual melhorada.


```python
# Aumentando o tamanho do gráfico com base no parâmetro "aspect" que por defeaut vem igual à 1, vamos passar o valor de 2

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2);
```

```python
# Ordenando os valores
''' No parâmetro "order" estamos pegando os indices da contagem de cada idioma e atribuindo para que o mesmo siga a ordem
do maior para o menor assim como é feito em "value_counts()"'''

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2,
            order=total_de_lingua_de_outros_idiomas.index);
```

```python
# Colando uma cor sequencial, ou seja da maior intensidade para a menor

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2,
            palette="mako",
            order=total_de_lingua_de_outros_idiomas.index);
```

# **Média, mediana, desvio padrão, boxplot, histograma, tendência central e dispersão, como tudo isso se encaixa**

Agora vamos fazer análise de dois filmes, lembrando que importamos o arquivo "movies.csv", vamos fazer a visualização e selecionar dois filmes.

```python
# Visualizando os dois primeiros filmes que vamos trabalhar

filmes.head(2)
```

```python
# Visualizando as duas primeiras notas dos dois primeiros filmes

notas.head(2)
```

```python
# Fazendo a contagem quantidade de notas de cada um dos filmes

# Pesquisando as notas do filme Toy Story pelo indice
notas_filme_toy_story = notas.query("filmeID == 1")

# Pesquisando as notas do filme Jumanji pelo indice
notas_filme_jumanji = notas.query("filmeID == 2")

# Visualizando a quantidade de notas de cada
print("Notas Filme Toy Story:", len(notas_filme_toy_story),
      "\nNotas Filme Jumanji:", len(notas_filme_jumanji))
```

Comparando a média dos dois filmes!

```python
# Média filme Toy Story

media_toy_story = notas_filme_toy_story["nota"].mean()
print(f"Nota Média Toy Story: {media_toy_story:.2f}")
```

```python
# Média filme Jumanji

media_jumanji = notas_filme_jumanji["nota"].mean()
print(f"Nota Média Jumanji: {media_jumanji:.2f}")
```

Utilizando tanto a média quanto a mediana temos uma visualização de onde os nossos dados estão se centralizando, porém fazendo dessa forma, perdemos muito de informações para saber como esses dados se alternam, dando pra gente apenas uma resumo de uma única tendência fazendo com que perdemos a informação de distribuição dessas notas.

```python
# Visualizando a distribuição dessas notas utilizando o boxplot do Seaborn
# Filme Toy Story

sns.boxplot(x=notas_filme_toy_story["nota"]);
```

```python
# Filme Jumanji

sns.boxplot(x=notas_filme_jumanji["nota"]);
```

Aqui podemos ver que de fato onde estão a média dessas notas, porém podemos visualizar os pontos fora da curva e também a forma de como isto esta distribuido.

Certo, mais e se quiséssemos fazer essa visualização junta ? No sentido de plotar mais de uma filme em um mesmo boxplot ?

É possível, porém a visualização é mais para análises específicas onde se pode tirar proveito do conteudo delas, mas não é indicado.

```python
# Vamos plotar os 5 primeiros filmes e vamos pegar as notas dos filmes que estão dentro de cada indice indicado no query

sns.boxplot(x="filmeID", y="nota", data = notas.query("filmeID in [1,2,3,4,5]"));
```

É visto que nos dados 3, 4 e 5, não é visto a linha de média no boxplot, entretanto ela está ali, porém a visualização desta forma dificulta identificar a linha (por isso nao é tão indicado a plotagem neste formato de múltiplos valores)

Sabemos que existe uma maneira numérica de apontar as medidas centrais de um conjunto (mediana, média, e por aí vai), mas será que também existe uma maneira numérica de averiguar quão distantes os outros pontos estão dessas medidas?

Essa medida existe e é chamado de **desvio padrão** (ou standard deviation), um parâmetro que indica o quanto os dados do conjunto fogem de uma tendência central.



```python
# Visualizando o desvio padrão das notas dos dois filmes (Toy Story e Jumanji)

desvio_padrao_toy_story = notas_filme_toy_story["nota"].std()
desvio_padrao_jumanji = notas_filme_jumanji["nota"].std()

print(
    f"Desvio Padrão Toy Story: {desvio_padrao_toy_story:.2f} \nDesvio Padrão Jumanji: {desvio_padrao_jumanji:.2f}",
)
```

Usamos isso para entender a tendência central (encontrando a média e a mediana) ou usamos o desvio padrão para entender a distribuição de como esses dados estão.
