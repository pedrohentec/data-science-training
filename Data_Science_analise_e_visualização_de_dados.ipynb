Acima temos a visualiza√ß√£o de uma Series, por mais semelhante que seja a visualiza√ß√£o de uma tabela, nessa situa√ß√£o temos os indices `(index)` ou seja, quando possamos que o valor de `en` √© igual a `4505`, temos como se o `en` fosse o `index` e o `4505` fosse o `valor`. Por exemplo:

Com essa visualiza√ß√£o acima podemos ter como afirma√ß√£o que a visualiza√ß√£o inicial √© uma s√©rie. Por√©m caso queiramos ter como visualiza√ß√£o em DataFrame, podemos utilizar a fun√ß√£o `to_frame()`:

```python
# Renomeando as colunas

contagem_de_lingua.columns = ["original_language", "total"]
contagem_de_lingua.head()
```

# **Passando uma mensagem atraveÃÅs de visualizacÃßaÃÉo**

Quando estamos trabalhando na √°rea de dados, temos a miss√£o de transmitir uma "mensagem" atrav√©s das an√°lises, seja com gr√°ficos ou at√© mesmo com artigos explicativos.

Nessa situa√ß√£o onde estamos aprendendo sobre plotagens, temos como miss√£o entender e enviar uma mensagem atrav√©s daquilo que estamos analisando, ou seja, decidir que gr√°ficos seria o mais recomendado, legendas, tamanho, tipo de dados, e etc.

Com isso, vamos entender um pouco sobre qual a melhor recomenda√ß√£o para a nossa situa√ß√£o que se enquadra em uma situa√ß√£o real.

```python
# Testando plotagem de "Pizza" (torta em Ingl√™s)
# Obs.: Esse tipo de gr√°fico n√£o possui no Seaborn, utilizaremos o Matplotlib

plt.pie(contagem_de_lingua["total"], labels=contagem_de_lingua["original_language"]);
```

Como podemos observar o gr√°fico de formato de Pizza (pie (torta em ingl√™s)), n√£o √© muito recomendado, pois a olho nu, fica "quase imposs√≠vel" de fazer uma "leitura" dessa plotagem.

Tivemos o gr√°fico acima e o gr√°fico de barras, por√©m tanto um como o outro n√£o servem para transmitir uma mensagem nit√≠da, que √© a compara√ß√£o de quantidade de idiomas dos filmes, nessa situa√ß√£o estamos apenas exercitando algumas pr√°ticas de plotagem, por√©m quando temos como definido o que realmente desejamos transmitir, fica mais "fac√≠l", porque podemos pesquisar ou encontrar modelos de plotagem que se adequariam melhor para a situa√ß√£o.

Por√©m como dito acima, como podemos entender melhor a mensagem que queremos transmitir ? Nessa situa√ß√£o fica evidente que a l√≠ngua inglesa √© disparada a que mais esta nos filmes, com isso vamos fazer a comparada dela com as outras linguas.

```python
# Pegando o total de valores das lingua na coluna "original_language"
total_de_lingas = filmes_tmdb["original_language"].value_counts()

# Somando o total geral das l√≠guas
total_geral = total_de_lingas.sum()

# Coletando o total de valores da lingua inglesa
total_ingles = total_de_lingas.loc["en"]

# Coletando o resto dos idiomas
total_do_resto = total_geral - total_ingles

# Printando a quantidade de valores de ingles e do resto dos idiomas
print(f"Valor Total Ingl√™s = {total_ingles}",
      f"\nValor Outros Idiomas = {total_do_resto}")
```

```python
# Criando um dataframe com base nas informa√ß√µes acima

# Colunas = "lingua" e "total"
# Coluna "lingua" = "ingles" e "outros"
# Coluna "total" = os valores acima de (seguindo a ordem das colunas acima) "total_ingles" e "total_do_resto"


dados_linguas = {
    "lingua" : ["ingl√™s", "outros"],
    "total" : [total_ingles, total_do_resto]
}

dados_linguas
```

```python
# Criando o DataFrame com base na constru√ß√£o acima

dados_linguas = pd.DataFrame(dados_linguas)
dados_linguas
```

```python
# Plotagem em barras usando o seaborn

sns.barplot(data=dados_linguas, x="lingua", y="total");
```

Com isso podemos finalizar a mensagem que queriamos transmitir, qual seja a compara√ß√£o de quantidade de filmes de lingua inglesa com os demais idiomas.

# **Visualizando a categoria "Outros"**

Vimos a quest√£o da compara√ß√£o entre o idioma ingl√™s com os outros idiomas, por√©m se quis√©ssemos entender como isto esta sendo distribuido, como poderiamos fazer isto ?

- Primeiro vamos selecionar os outros idiomas que sejam diferentes de ingl√™s

```python
# Selecionando idiomas diferentes de ingl√™s

filmes_sem_lingua_em_ingles = filmes_tmdb.query("original_language != 'en'")
filmes_sem_lingua_em_ingles.head()
```

```python
# Fazendo a contagem de valores de outros idiomas

total_de_lingua_de_outros_idiomas = filmes_sem_lingua_em_ingles["original_language"].value_counts()
total_de_lingua_de_outros_idiomas
```

Depois de separmos quais os dados que vamos trabalhar para plotar ou seja, os filmes SEM o idioma ingl√™s, vamos agora fazer a plotagem do mesmo.

```python
sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count");
```

# **Refinando visualiza√ß√µes**

Com base no gr√°fico acima, vamos melhorar a visualiza√ß√£o dele para termos uma an√°lise visual melhorada.


```python
# Aumentando o tamanho do gr√°fico com base no par√¢metro "aspect" que por defeaut vem igual √† 1, vamos passar o valor de 2

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2);
```

```python
# Ordenando os valores
''' No par√¢metro "order" estamos pegando os indices da contagem de cada idioma e atribuindo para que o mesmo siga a ordem
do maior para o menor assim como √© feito em "value_counts()"'''

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2,
            order=total_de_lingua_de_outros_idiomas.index);
```

```python
# Colando uma cor sequencial, ou seja da maior intensidade para a menor

sns.catplot(data=filmes_sem_lingua_em_ingles, x="original_language", kind="count",
            aspect=2,
            palette="mako",
            order=total_de_lingua_de_outros_idiomas.index);
```

# **M√©dia, mediana, desvio padr√£o, boxplot, histograma, tend√™ncia central e dispers√£o, como tudo isso se encaixa**

Agora vamos fazer an√°lise de dois filmes, lembrando que importamos o arquivo "movies.csv", vamos fazer a visualiza√ß√£o e selecionar dois filmes.

```python
# Visualizando os dois primeiros filmes que vamos trabalhar

filmes.head(2)
```

```python
# Visualizando as duas primeiras notas dos dois primeiros filmes

notas.head(2)
```

```python
# Fazendo a contagem quantidade de notas de cada um dos filmes

# Pesquisando as notas do filme Toy Story pelo indice
notas_filme_toy_story = notas.query("filmeID == 1")

# Pesquisando as notas do filme Jumanji pelo indice
notas_filme_jumanji = notas.query("filmeID == 2")

# Visualizando a quantidade de notas de cada
print("Notas Filme Toy Story:", len(notas_filme_toy_story),
      "\nNotas Filme Jumanji:", len(notas_filme_jumanji))
```

Comparando a m√©dia dos dois filmes!

```python
# M√©dia filme Toy Story

media_toy_story = notas_filme_toy_story["nota"].mean()
print(f"Nota M√©dia Toy Story: {media_toy_story:.2f}")
```

```python
# M√©dia filme Jumanji

media_jumanji = notas_filme_jumanji["nota"].mean()
print(f"Nota M√©dia Jumanji: {media_jumanji:.2f}")
```

Utilizando tanto a m√©dia quanto a mediana temos uma visualiza√ß√£o de onde os nossos dados est√£o se centralizando, por√©m fazendo dessa forma, perdemos muito de informa√ß√µes para saber como esses dados se alternam, dando pra gente apenas uma resumo de uma √∫nica tend√™ncia fazendo com que perdemos a informa√ß√£o de distribui√ß√£o dessas notas.

```python
# Visualizando a distribui√ß√£o dessas notas utilizando o boxplot do Seaborn
# Filme Toy Story

sns.boxplot(x=notas_filme_toy_story["nota"]);
```

```python
# Filme Jumanji

sns.boxplot(x=notas_filme_jumanji["nota"]);
```

Aqui podemos ver que de fato onde est√£o a m√©dia dessas notas, por√©m podemos visualizar os pontos fora da curva e tamb√©m a forma de como isto esta distribuido.

Certo, mais e se quis√©ssemos fazer essa visualiza√ß√£o junta ? No sentido de plotar mais de uma filme em um mesmo boxplot ?

√â poss√≠vel, por√©m a visualiza√ß√£o √© mais para an√°lises espec√≠ficas onde se pode tirar proveito do conteudo delas, mas n√£o √© indicado.

```python
# Vamos plotar os 5 primeiros filmes e vamos pegar as notas dos filmes que est√£o dentro de cada indice indicado no query

sns.boxplot(x="filmeID", y="nota", data = notas.query("filmeID in [1,2,3,4,5]"));
```

√â visto que nos dados 3, 4 e 5, n√£o √© visto a linha de m√©dia no boxplot, entretanto ela est√° ali, por√©m a visualiza√ß√£o desta forma dificulta identificar a linha (por isso nao √© t√£o indicado a plotagem neste formato de m√∫ltiplos valores)

Sabemos que existe uma maneira num√©rica de apontar as medidas centrais de um conjunto (mediana, m√©dia, e por a√≠ vai), mas ser√° que tamb√©m existe uma maneira num√©rica de averiguar qu√£o distantes os outros pontos est√£o dessas medidas?

Essa medida existe e √© chamado de **desvio padr√£o** (ou standard deviation), um par√¢metro que indica o quanto os dados do conjunto fogem de uma tend√™ncia central.



```python
# Visualizando o desvio padr√£o das notas dos dois filmes (Toy Story e Jumanji)

desvio_padrao_toy_story = notas_filme_toy_story["nota"].std()
desvio_padrao_jumanji = notas_filme_jumanji["nota"].std()

print(
    f"Desvio Padr√£o Toy Story: {desvio_padrao_toy_story:.2f} \nDesvio Padr√£o Jumanji: {desvio_padrao_jumanji:.2f}",
)
```

Usamos isso para entender a tend√™ncia central (encontrando a m√©dia e a mediana) ou usamos o desvio padr√£o para entender a distribui√ß√£o de como esses dados est√£o.
